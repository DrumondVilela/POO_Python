# Classe Bomba de Combustível:
## Faça um programa que:
### a. Possua uma classe chamada bombaCombustivel, com no mínimo esses atributos:
- TipoCombustivel;
- valorLitro;
- quantidadeCombustivel.

### b. Possua no mínimo esses métodos:
- abastecerPorValor( ) – método onde é informado o valor a ser abastecido e mostra a quantidade de litros que foi colocada no veículo
- abastecerPorLitro( ) – método onde é informado a quantidade em litros de combustível e mostra o valor a ser pago pelo cliente.
- alterarValor( ) – altera o valor do litro do combustível.iv.alterarCombustivel( ) –altera o tipo do combustível.
- alterarQuantidadeCombustivel( ) – altera a quantidade de combustível restante na bomba

```python
# gasolinaComum = 6.89
# gasolinaAditivada = 7.05
# gas = 5.96
# etanol = 6.32

# Criação da classe:
class bombaCombustivel():

    # Define os atributos da classe
    def __init__(self,tipoCombustivel, valorLitro, qtdCombustivel):
        self.tipoCombustivel = tipoCombustivel
        self.valorLitro = valorLitro
        self.qtdCombustivel = qtdCombustivel


    # Define os métodos da classe:
    def abastecerPorValor(self):
        while True:
            try:
                valor = float(input('Digite o valor que quer adicionar de combustível: '))
                break
            except:
                print('Valor inválido! Tente Novamente.')
        litros = valor/self.valorLitro
        self.qtdCombustivel = self.qtdCombustivel - litros
        print(f'Foram adicionados {litros:.2f}L de combustível ao veículo.')

    def abastecerPorLitro(self):
        while True:
            try:
                litros = float(input('Digite a quantidade de combustível que deseja adicionar: '))
                break
            except:
                print('[ERRO] Valor não aceitável! Por favor, tente novamente.')
        total = litros*self.valorLitro
        self.qtdCombustivel = self.qtdCombustivel - litros
        print(f'Total a pagar:R${total:.2f}')

    def alterarValor(self):
        while True:
            try:
                novoValor = float(input('Digite o novo valor de combustível: '))
                break
            except:
                print('[ERRO] Valor Inválido! Tente novamente.')
        self.valorLitro = novoValor
        print(f'Valor do combustível alterado para R${novoValor}.')

    def alterarCombustivel(self):
        while True:
            try:
                novoTipo = input('Digite o novo tipo de combustível: ')
                break
            except:
                print('[ERRO] Valor Inválido! Tente novamente.')
        self.tipoCombustivel = novoTipo
        print(f'Tipo de combustível alterado para {novoTipo}.')
        
    def verQtd(self):
        print(f'Há {self.qtdCombustivel:.3f}L de combustível na bomba.')


posto = bombaCombustivel('Gasolina Comum', 6.89,500)
opcoes = [1,2,3,4]
print('='*25+' Bem-Vindo(a)! '+'='*25)

while True:
    try:
        entrada = int(input('''
        [1] - Abastecer por valor.
        [2] - Abastecer por litro de combustível
        [3] - Alterar o combustível
        [4] - Alterar valor do combustível
        Resposta: '''))
        while entrada not in opcoes:
            print('[ERRO-01] VALOR INVÁLIDO. TENTE NOVAMENTE!')
            entrada = int(input('''
        [1] - Abastecer por valor.
        [2] - Abastecer por litro de combustível
        [3] - Alterar o combustível
        [4] - Alterar valor do combustível
        Resposta: '''))
        break
    except:
        print('[ERRO-02] VALOR INVÁLIDO. TENTE NOVAMENTE!')

if entrada == 1:
    posto.abastecerPorValor()
    posto.verQtd()
elif entrada == 2:
    posto.abastecerPorLitro()
    posto.verQtd()
elif entrada == 3:
    posto.alterarValor() 
    posto.verQtd()
elif entrada == 4:
    posto.alterarValor()
    posto.verQtd()
print('='*25 + ' Volte Sempre! ' + '='*25)


```
